```python
def calculate_average():
    num_st = int(input("Number Of Students:"))
    grades = []
    
    for i in range(num_st) :
        grade = float(input("Enter The Grade : "))
        grades.append(grade)

    if num_st == 0:
        print ("No Students")
        
    else:
        print(f"The Average is : {sum(grades)/num_st}")
    
calculate_average() 
```

```python
import os
import random

def thanos_project(base_dir, num_files):
    if not os.path.exists(base_dir):
        os.makedirs(base_dir)

    for i in range(num_files):
        file_name = "file_" + str(i) + ".txt"
        file_path = os.path.join(base_dir, file_name)
        with open(file_path, "w") as f:
            f.write("Thanos was here")

    files = os.listdir(base_dir)
    print("Number of files before deletion:", len(files))
    
    half = len(files) // 2
    random_files = random.sample(files, half)

    for file in random_files:
        os.remove(os.path.join(base_dir, file))

    final_files = os.listdir(base_dir)
    print("Number of files after deletion:", len(final_files))
```
